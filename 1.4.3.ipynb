# 1.4.3 Geometric Transformations of Images

#Scaling

```python
import cv2
import numpy as np
img = cv2.imread('./image/messi.jpg')
res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
#OR
height, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)
```

```python
#Translation
```

```python
import cv2
import numpy as np
img = cv2.imread('./image/messi.jpg',0)
rows,cols = img.shape
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Rotation
```

```python
img = cv2.imread('./image/messi.jpg',0)
rows,cols = img.shape
M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Affine Transformation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./image/dsu4.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[235,457],[238,269],[569,231]])
pts2 = np.float32([[235,457],[235,270],[570,270]])
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

# Perspective Transformation

```python
from matplotlib import pyplot as plt

img = cv2.imread('./image/dsu4.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[240,174],[571,99],[234,536],[568,607]])
pts2 = np.float32([[100,100],[200,100],[100,200],[200,200]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(300,300))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()

```
